```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

# Aplcando Metodo SVD a imagen

```python
plt.style.use("classic")
foto = Image.open("./imgs/pintura.jpg")
plt.imshow(foto)
```

```python
# convertimos la foto a escala de grises
foto_bn = foto.convert("LA")
# obtenemos un vector con todos los elementos del canal rojo de la imagen
foto_bn_mat = np.array(list(foto_bn.getdata(band=0)), float)
# Le damos las dimensiones la la imagen a la lista obtenida
foto_bn_mat.shape = (foto_bn.size[1], foto_bn.size[0])

# mostramos nuestra imagen obtenida en escala de grises
plt.imshow(foto_bn)
```

```python
# Aplicamos metodo SVD a foto en escala de grises
U, D, V = np.linalg.svd(foto_bn_mat)
```

```python
# Comprobamos dimensiones
print(foto_bn_mat.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

# Reconstruimos imagen probando can varias cantidades de los valores singulares

```python
# Intento 1: Intentamos reconstruir la imagen a partir de 1 valor singular
# la primera columna de U, el primer elemento de D y la primera fila de V

foto_recovered = np.matrix(U[:,:1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(foto_recovered, cmap="gray")
```

```python
# Intento 2: Intentamos reconstruir la imagen a partir de 5 valores singulares
# N filas de U, N elementos de D y N filas de V
N = 5
foto_recovered = np.matrix(U[:,:N]) * np.diag(D[:N]) * np.matrix(V[:N,:])
plt.imshow(foto_recovered, cmap="gray")

titulo = "valores singulares = %s" % N
plt.title(titulo)
plt.show()
```
