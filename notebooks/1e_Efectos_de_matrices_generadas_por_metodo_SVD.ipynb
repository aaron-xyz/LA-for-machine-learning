# Equivalencia entre la matriz A y las matrices U, D V producto del metodo SVD

Dada una matriz $A$, al aplicarsele el metodo SVD (Singular Value Decomposition) generamos 3 matrices diferentes, las matrices U, D y V. Estas matrices generan una rotacion, un cambio de escala y otra rotacion respectivamente. Al final, si aplicamos estas matrices en el orden UDV obtendremos la misma transformacion que la matriz A genera.

Efecto de la descomposición de una matriz en vectores singulares:
- V → rota el espacio
- D → Escala el espacio (agranda o encoje los vectores, también puede cambiar su sentido)
- U → rota nuevamente los vectores

Nota:
La descomposición por valores singulares tiene efectos similares:

- autovectores → rota el espacio
- diag(autovalores) → escala el espacio
- inv(autovectores) → rota el espacio

```python
# Calculemos las matrices UDV de A usando el metodo SVD
U, D, V = np.linalg.svd(A)

print("Matriz U:")
print(U)
print("Matriz D:")
print(np.diag(D))
print("Matriz V:")
print(V)
```

```python
# Ahora veamos los efectos de cada una de estas 3 matrices
# Empezamos por visualizar el circulo unitario
print("Circulo Unitario")
graficarMatriz(np.eye(2))

plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()

# Graficamos el efecto de la matriz V
print("Primera Rotacion: V")
graficarMatriz(V)

plt.xlim(-2,2)
plt.ylim(-2,2)
plt.show()

# Graficamos el efecto de la matriz D
print("Escala: D")
graficarMatriz(np.diag(D).dot(V))

plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()

# Graficamos el efecto de la matriz U
print("Segunda Rotacion: U")
graficarMatriz(U.dot(np.diag(D).dot(V)))

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# comparamos con la transformacion generada por la matriz A para comprobar que el resultado final es el mismo
print("Circulo Unitario transformado")
graficarMatriz(A)

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
